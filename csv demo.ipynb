```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import os
import glob
```

```python
def return_lever_press(file):
    df = pd.read_csv(file, skiprows=3)
    value = df['total_lever_press'].values[0]
    
    return value
```

```python
files = glob.glob('/Users/davidprotter/Downloads/reanalyzed_files/*_summary.csv')
```

```python
files_to_analyze = [f for f in sorted(files) if '4049' in f]
files_to_analyze_sublist = files_to_analyze[0:4]

output_4049 = []

for file in files_to_analyze_sublist:
    output_4049 += [return_lever_press(file)]
    
files_to_analyze = [f for f in sorted(files) if '3946' in f]
files_to_analyze_sublist = files_to_analyze[0:4]

output_3946 = []

for file in files_to_analyze_sublist:
    output_3946 += [return_lever_press(file)]
```

```python
fig, axs = plt.subplots(ncols=2, sharex = True, sharey =True)
axs[0].plot(output_3946, label = '3946')
axs[1].plot(output_4049, label = '4049')
```
